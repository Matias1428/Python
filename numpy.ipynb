## Modulos

Antes de hablar de `numpy`, hablemos de los módulos de Python.  
Los módulos son librerías de código, archivos conteniendo variables, funciones, clases y estructuras de datos que podemos importar y usar en nuestras aplicaciones.  
   
Python tiene varios [módulos que vienen incluídos por default](https://docs.python.org/3/library/).  

Para los siguientes ejemplos usaremos el módulo [math](https://docs.python.org/3/library/math.html), pero invito a que se lea la documentación de los módulos que parezcan interesantes.


Para importar un modulo, usamos la palabra clave `import`

```python
import math
```

```python
e = math.exp(1)
print(e)
```

Es posible importar solo una parte de un modulo usando:  
 `from` \<modulo\> `import`  \<objeto\>

```python
from math import sqrt
```

```python
print(sqrt(9))
print(sqrt(16))
```

Es posible importar multiples objetos simultaneamente

```python
from math import cos, sin, pi

print(cos(0), sin(0))
print(sin(pi), cos(pi))
```

La palabra clave `as` nos permite ponerle un alias a los modulos o librerias importados.

```python
from math import inf as infinito

print(infinito)
print(infinito + 1)
```

Los arrays son la estructura de datos central de la biblioteca NumPy. Tiene una cuadrícula de elementos que se pueden indexar de varias maneras. Los elementos son todos del mismo tipo, referido como el dtype del array.

Hay momentos en los que es posible que desee realizar una operación entre un array y un solo número (también llamada operación entre un vector y un escalar).

#### Indexación de arrays (Acceder a elmentos y subconjuntos)

Uno puede hacer uso de los indices y de los 'slices' `n:m` de la misma manera que lo hacemos con las listas

numpy-array-slice.png

Además, es extremadamente simple hacer selección condicional de elementos en un array.

```python
a = np.arange(10)
```

```python
print(a < 5)
```

```python
# Nota que le estamos pasando al array, el vector de booleanos de la celda 
#   pasada y nos regresa los elementos donde su indice fue True
print(a[a < 5])
```

```python
c = a[(a > 2) & (a < 7)]
print(c)
```

#### `np.where()`

np.where nos da herramientas para realizar operaciones condicionales sobre nuestros arrays.

```python
arr = np.random.randint(10, size=10)
print(arr)
```

```python
np.where(arr > 5)
```

```python
np.where(arr > 5, True, False)
```

```python
np.where(arr > 5, "mayor a cinco", "menor o igual a cinco")
```

```python
np.where(arr > 5, arr, -arr)
```

### Matrices

Si pensamos en matrices como vectores de dos dimensiones, las matrices en numpy no es otra cosas que arrays anidados representando dos dimensiones.

#### Crear Matrices en Numpy

Le podemos pasar una lista de listas para crear un arrat en 2-D (i.e. una matriz)

numpy-array-create-2d.png

Las mismas funciones para crear arrays, funcionan para crear matrices. La unica diferencia es que pasamos como argumento la dimensionalidad de las matrices dentro de una tupla `np.foo((n_filas, n_columnas))`

numpy-matrix-ones-zeros-random.png

#### Aritmética

Las operaciones aritmeticas para matrices de misma dimensionalidad funcionan de la misma manera que con los vectores.

```python
data = np.array([[1, 2], [3, 4]])
ones = np.ones((2, 2))
```

numpy-matrix-arithmetic.png

El broadcasting en las matrices funciona de manera similar.


*   Matriz con escalar: se aplica uno a uno la operacion a cada elemento
*   Matriz con vector con filas del mismo tamaño: se aplica a cada fila la suma como si fueran arrays del mismo tamaño.
*   Matriz con vector con columnas del mismo tama{o: se aplica a cada columna la suma como si fueran arrays del mismo tamaño.



numpy-matrix-broadcast.png

```python
data = np.arange(1, 7).reshape(3, 2)
ones_row = np.ones((1, 2))
ones_col = np.ones((3, 1))
```

```python
data + 1.0
```

```python
data + ones_col
```

```python
data + ones_row
```
